```python
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
from PIL import Image, ImageTk
import tkinterdnd2 as tkdnd  # Новый импорт!

def on_drop(event):
    """Обработка перетаскивания файлов"""
    files = root.tk.splitlist(event.data)
    
    # Фильтруем изображения
    image_files = [f for f in files 
                   if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
    
    if image_files:
        load_to_slots(image_files[:3])
        status_var.set(f"Перетащено {len(image_files)} файлов")
    else:
        messagebox.showwarning("Внимание", "Перетащите файлы изображений!")

def load_to_slots(file_paths):
    """Загрузка изображений в слоты"""
    for i, file_path in enumerate(file_paths):
        try:
            img = Image.open(file_path)
            img.thumbnail((200, 200))
            photo = ImageTk.PhotoImage(img)
            
            image_labels[i].configure(image=photo, text="")
            image_labels[i].image = photo
        except Exception as e:
            messagebox.showerror("Ошибка", f"Ошибка загрузки: {e}")

def load_images():
    files = filedialog.askopenfilenames(
        title="Выберите изображения",
        filetypes=[("Изображения", "*.png *.jpg *.jpeg *.gif *.bmp")]
    )
    if files:
        load_to_slots(files[:3])
        status_var.set(f"Загружено {min(3, len(files))} изображений")

def show_info():
    messagebox.showinfo("О приложении", 
                       "Просмотр изображений v1.0\n\n"
                       "• Загрузка через кнопку\n"
                       "• Drag and Drop\n"
                       "• Просмотр до 3 изображений")

def clear_images():
    for label in image_labels:
        label.configure(image="", text="Нет\nизображения")
        label.image = None
    status_var.set("Изображения очищены")

# Используем TkinterDnD.Tk вместо tk.Tk!
root = tkdnd.Tk()
root.title("Просмотр изображений")
root.geometry("800x500")

# Регистрируем окно для приема перетаскиваемых файлов
root.drop_target_register(tkdnd.DND_FILES)
root.dnd_bind('<<Drop>>', on_drop)

main_frame = ttk.Frame(root, padding="20")
main_frame.pack(fill=tk.BOTH, expand=True)

# Область для drag and drop
drop_frame = ttk.LabelFrame(main_frame, text="Перетащите файлы сюда", padding="10")
drop_frame.pack(fill=tk.X, pady=(0, 10))

drop_label = ttk.Label(drop_frame, 
                      text="ПЕРЕТАЩИТЕ ИЗОБРАЖЕНИЯ СЮДА",
                      background="lightblue", padding="15",
                      font=("Arial", 11, "bold"), anchor="center")
drop_label.pack(fill=tk.X)

# Фрейм для изображений
images_frame = ttk.LabelFrame(main_frame, text="Просмотр", padding="10")
images_frame.pack(fill=tk.BOTH, expand=True, pady=10)

image_labels = []

for i in range(3):
    container = ttk.Frame(images_frame, relief="solid", borderwidth=1)
    container.pack(side=tk.LEFT, padx=10, expand=True, fill=tk.BOTH)
    
    ttk.Label(container, text=f"Изображение {i+1}").pack(pady=5)
    
    img_label = ttk.Label(container, text="Нет\nизображения", 
                          background="lightgray", anchor="center")
    img_label.pack(padx=10, pady=10, fill=tk.BOTH, expand=True)
    
    image_labels.append(img_label)

# Кнопки
buttons_frame = ttk.Frame(main_frame)
buttons_frame.pack(pady=20)

ttk.Button(buttons_frame, text="Загрузить", 
          command=load_images, width=15).pack(side=tk.LEFT, padx=5)
ttk.Button(buttons_frame, text="Информация", 
          command=show_info, width=15).pack(side=tk.LEFT, padx=5)
ttk.Button(buttons_frame, text="Очистить", 
          command=clear_images, width=15).pack(side=tk.LEFT, padx=5)

status_var = tk.StringVar(value="Готов к работе. Перетащите файлы или нажмите 'Загрузить'")
ttk.Label(main_frame, textvariable=status_var, 
         relief=tk.SUNKEN).pack(fill=tk.X, pady=(10, 0))

root.mainloop()
```

Автоматически сохранять загруженные изображения

```python
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
from PIL import Image, ImageTk
import tkinterdnd2 as tkdnd
import os
import shutil

class ImageApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Просмотр изображений")
        self.root.geometry("800x500")
        
        # Создаем папку data
        self.data_dir = "data"
        if not os.path.exists(self.data_dir):
            os.makedirs(self.data_dir)
            print(f"Создана папка: {self.data_dir}")
        
        self.image_labels = []
        self.setup_ui()
        
        # Регистрация Drag and Drop
        self.root.drop_target_register(tkdnd.DND_FILES)
        self.root.dnd_bind('<<Drop>>', self.on_drop)
    
    def setup_ui(self):
        main_frame = ttk.Frame(self.root, padding="20")
        main_frame.pack(fill=tk.BOTH, expand=True)
        
        # Область drag and drop
        drop_frame = ttk.LabelFrame(main_frame, text="Перетащите файлы сюда", padding="10")
        drop_frame.pack(fill=tk.X, pady=(0, 10))
        
        ttk.Label(drop_frame, 
                 text="ПЕРЕТАЩИТЕ ИЗОБРАЖЕНИЯ СЮДА",
                 background="lightblue", padding="15",
                 font=("Arial", 11, "bold"), anchor="center").pack(fill=tk.X)
        
        # Фрейм для изображений
        images_frame = ttk.LabelFrame(main_frame, text="Просмотр", padding="10")
        images_frame.pack(fill=tk.BOTH, expand=True)
        
        for i in range(3):
            container = ttk.Frame(images_frame, relief="solid", borderwidth=1)
            container.pack(side=tk.LEFT, padx=10, expand=True, fill=tk.BOTH)
            
            ttk.Label(container, text=f"Изображение {i+1}").pack(pady=5)
            
            img_label = ttk.Label(container, text="Нет\nизображения", 
                                 background="lightgray", anchor="center")
            img_label.pack(padx=10, pady=10, fill=tk.BOTH, expand=True)
            
            self.image_labels.append(img_label)
        
        # Кнопки
        buttons_frame = ttk.Frame(main_frame)
        buttons_frame.pack(pady=20)
        
        ttk.Button(buttons_frame, text="Загрузить", 
                  command=self.load_images, width=15).pack(side=tk.LEFT, padx=5)
        ttk.Button(buttons_frame, text="Информация", 
                  command=self.show_info, width=15).pack(side=tk.LEFT, padx=5)
        ttk.Button(buttons_frame, text="Очистить", 
                  command=self.clear_images, width=15).pack(side=tk.LEFT, padx=5)
        
        self.status_var = tk.StringVar(value="Готов к работе")
        ttk.Label(main_frame, textvariable=self.status_var, 
                 relief=tk.SUNKEN).pack(fill=tk.X, pady=(10, 0))
    
    def save_to_data_folder(self, file_paths):
        """Сохранение файлов в папку data"""
        saved_count = 0
        for file_path in file_paths:
            try:
                filename = os.path.basename(file_path)
                dest_path = os.path.join(self.data_dir, filename)
                
                # Если файл существует, добавляем номер
                counter = 1
                base_name, ext = os.path.splitext(filename)
                while os.path.exists(dest_path):
                    new_name = f"{base_name}_{counter}{ext}"
                    dest_path = os.path.join(self.data_dir, new_name)
                    counter += 1
                
                shutil.copy2(file_path, dest_path)
                saved_count += 1
                print(f"Сохранен: {dest_path}")
                
            except Exception as e:
                messagebox.showerror("Ошибка", f"Ошибка сохранения: {e}")
        
        return saved_count
    
    def process_files(self, files):
        """Обработка файлов"""
        # Фильтруем изображения
        image_files = [f for f in files 
                      if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
        
        if not image_files:
            messagebox.showwarning("Внимание", "Нет файлов изображений!")
            return
        
        # Отображаем первые 3
        for i, file_path in enumerate(image_files[:3]):
            try:
                img = Image.open(file_path)
                img.thumbnail((200, 200))
                photo = ImageTk.PhotoImage(img)
                
                self.image_labels[i].configure(image=photo, text="")
                self.image_labels[i].image = photo
            except Exception as e:
                messagebox.showerror("Ошибка", f"Ошибка отображения: {e}")
        
        # Сохраняем ВСЕ изображения
        saved = self.save_to_data_folder(image_files)
        self.status_var.set(f"Отображено {min(3, len(image_files))}, сохранено {saved} файлов")
    
    def on_drop(self, event):
        files = self.root.tk.splitlist(event.data)
        self.process_files(files)
    
    def load_images(self):
        files = filedialog.askopenfilenames(
            title="Выберите изображения",
            filetypes=[("Изображения", "*.png *.jpg *.jpeg *.gif *.bmp")]
        )
        if files:
            self.process_files(list(files))
    
    def show_info(self):
        info_text = f"""Просмотр изображений v1.0

Возможности:
• Загрузка через диалог
• Drag and Drop
• Отображение до 3 изображений
• Сохранение в папку '{self.data_dir}'

Поддерживаемые форматы:
PNG, JPG, JPEG, GIF, BMP"""
        
        messagebox.showinfo("О приложении", info_text)
    
    def clear_images(self):
        for label in self.image_labels:
            label.configure(image="", text="Нет\nизображения")
            label.image = None
        self.status_var.set("Изображения очищены")

if __name__ == "__main__":
    root = tkdnd.Tk()
    app = ImageApp(root)
    root.mainloop()
```

Сделать интерфейс адаптивным и центрированным

```python
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
from PIL import Image, ImageTk
import tkinterdnd2 as tkdnd
import os
import shutil

class ImageApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Просмотр изображений")
        
        # Центрируем окно
        self.center_window(900, 600)
        
        # Настройка адаптивности
        self.root.grid_rowconfigure(0, weight=1)
        self.root.grid_columnconfigure(0, weight=1)
        
        self.data_dir = "data"
        if not os.path.exists(self.data_dir):
            os.makedirs(self.data_dir)
        
        self.image_labels = []
        self.setup_ui()
        
        self.root.drop_target_register(tkdnd.DND_FILES)
        self.root.dnd_bind('<<Drop>>', self.on_drop)
    
    def center_window(self, width, height):
        """Центрирование окна на экране"""
        x = (self.root.winfo_screenwidth() // 2) - (width // 2)
        y = (self.root.winfo_screenheight() // 2) - (height // 2)
        self.root.geometry(f'{width}x{height}+{x}+{y}')
    
    def setup_ui(self):
        # Главный фрейм с grid для лучшего контроля
        main_frame = ttk.Frame(self.root, padding="20")
        main_frame.grid(row=0, column=0, sticky="nsew")
        
        # Настройка весов для адаптивности
        main_frame.grid_rowconfigure(0, weight=1)  # Изображения
        main_frame.grid_rowconfigure(1, weight=1)  # Drop zone
        main_frame.grid_rowconfigure(2, weight=0)  # Кнопки
        main_frame.grid_rowconfigure(3, weight=0)  # Статус
        main_frame.grid_columnconfigure(0, weight=1)
        
        # Drop zone
        drop_frame = ttk.LabelFrame(main_frame, text="Drag & Drop зона", padding="10")
        drop_frame.grid(row=0, column=0, pady=(0, 10), sticky="nsew")
        
        drop_label = ttk.Label(drop_frame, 
                              text=" ПЕРЕТАЩИТЕ ФАЙЛЫ ИЗОБРАЖЕНИЙ СЮДА"
                                   "PNG, JPG, JPEG, GIF, BMP",
                              background=#E1F0FF", padding="20",
                              font=("Arial", 11), anchor="center", justify="center")
        drop_label.pack(fill=tk.BOTH, expand=True)
        
        # Фрейм изображений
        images_frame = ttk.LabelFrame(main_frame, text="Просмотр изображений", padding="15")
        images_frame.grid(row=1, column=0, pady=10, sticky="nsew")
        
        # Равномерное распределение колонок
        for i in range(3):
            images_frame.grid_columnconfigure(i, weight=1)
        images_frame.grid_rowconfigure(0, weight=1)
        
        for i in range(3):
            container = ttk.Frame(images_frame)
            container.grid(row=0, column=i, padx=10, pady=5, sticky="nsew")
            
            container.grid_rowconfigure(0, weight=0)
            container.grid_rowconfigure(1, weight=1)
            container.grid_columnconfigure(0, weight=1)
            
            frame = ttk.Frame(container, relief="solid", borderwidth=2)
            frame.grid(row=0, column=0, rowspan=2, sticky="nsew", padx=5, pady=5)
            
            frame.grid_rowconfigure(0, weight=0)
            frame.grid_rowconfigure(1, weight=1)
            frame.grid_columnconfigure(0, weight=1)
            
            ttk.Label(frame, text=f"Изображение {i+1}", 
                     font=("Arial", 10, "bold")).grid(row=0, column=0, pady=5)
            
            img_label = ttk.Label(frame, text="Нет\nизображения", 
                                 background="#F0F0F0", anchor="center")
            img_label.grid(row=1, column=0, padx=10, pady=10, sticky="nsew")
            
            self.image_labels.append(img_label)
        
        # Кнопки с центрированием
        buttons_frame = ttk.Frame(main_frame)
        buttons_frame.grid(row=2, column=0, pady=20)
        
        # Центрируем кнопки
        buttons_frame.grid_columnconfigure(0, weight=1)
        buttons_frame.grid_columnconfigure(4, weight=1)
        
        ttk.Button(buttons_frame, text="Загрузить", 
                  command=self.load_images, width=18).grid(row=0, column=1, padx=5)
        ttk.Button(buttons_frame, text="Информация", 
                  command=self.show_info, width=15).grid(row=0, column=2, padx=5)
        ttk.Button(buttons_frame, text="Очистить", 
                  command=self.clear_images, width=15).grid(row=0, column=3, padx=5)
        
        # Статус бар
        self.status_var = tk.StringVar(value="Готов к работе")
        status_bar = ttk.Label(main_frame, textvariable=self.status_var, 
                              relief=tk.SUNKEN, anchor=tk.W, padding=(5, 2))
        status_bar.grid(row=3, column=0, sticky="ew")
    
    def process_files(self, files):
        image_files = [f for f in files 
                      if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
        
        if not image_files:
            messagebox.showwarning("Внимание", "Выбранные файлы не являются изображениями!")
            return
        
        # Отображаем первые 3
        for i, file_path in enumerate(image_files[:3]):
            try:
                img = Image.open(file_path)
                
                # Адаптивный размер
                label_width = max(150, self.image_labels[i].winfo_width())
                label_height = max(150, self.image_labels[i].winfo_height())
                
                img.thumbnail((label_width, label_height), Image.Resampling.LANCZOS)
                photo = ImageTk.PhotoImage(img)
                
                self.image_labels[i].configure(image=photo, text="")
                self.image_labels[i].image = photo
            except Exception as e:
                messagebox.showerror("Ошибка", f"Не удалось загрузить изображение:\n{e}")
        
        # Сохраняем все
        saved = self.save_to_data_folder(image_files)
        self.status_var.set(f"Загружено {len(image_files)} файлов, сохранено в '{self.data_dir}'")
    
    def save_to_data_folder(self, file_paths):
        saved_count = 0
        for file_path in file_paths:
            try:
                filename = os.path.basename(file_path)
                dest_path = os.path.join(self.data_dir, filename)
                
                counter = 1
                base_name, ext = os.path.splitext(filename)
                while os.path.exists(dest_path):
                    dest_path = os.path.join(self.data_dir, f"{base_name}_{counter}{ext}")
                    counter += 1
                
                shutil.copy2(file_path, dest_path)
                saved_count += 1
            except Exception as e:
                print(f"Ошибка сохранения {filename}: {e}")
        
        return saved_count
    
    def on_drop(self, event):
        files = self.root.tk.splitlist(event.data)
        self.process_files(files)
    
    def load_images(self):
        files = filedialog.askopenfilenames(
            title="Выберите изображения",
            filetypes=[("Изображения", "*.png *.jpg *.jpeg *.gif *.bmp *.tiff")]
        )
        if files:
            self.process_files(list(files))
    
    def show_info(self):
        info_window = tk.Toplevel(self.root)
        info_window.title("О приложении")
        info_window.geometry("400x350")
        info_window.resizable(False, False)
        
        # Центрирование
        info_window.update_idletasks()
        x = info_window.winfo_screenwidth() // 2 - 200
        y = info_window.winfo_screenheight() // 2 - 175
        info_window.geometry(f"400x350+{x}+{y}")
        
        info_window.transient(self.root)
        info_window.grab_set()
        
        frame = ttk.Frame(info_window, padding="20")
        frame.pack(fill=tk.BOTH, expand=True)
        
        ttk.Label(frame, text="Просмотр изображений", 
                 font=("Arial", 14, "bold")).pack(pady=10)
        ttk.Label(frame, text="Версия 1.0").pack()
        
        ttk.Label(frame, text="\nВозможности:", 
                 font=("Arial", 11, "bold")).pack(anchor=tk.W, pady=(20, 5))
        
        features = """• Отображение до 3 изображений
• Drag & Drop загрузка
• Выбор файлов через диалог
• Автосохранение в папку 'data'
• Адаптивный интерфейс
• Поддержка популярных форматов"""
        
        ttk.Label(frame, text=features, justify=tk.LEFT).pack(anchor=tk.W)
        
        ttk.Button(frame, text="Закрыть", 
                  command=info_window.destroy, width=15).pack(pady=20)
    
    def clear_images(self):
        for label in self.image_labels:
            label.configure(image="", text="Нет\nизображения")
            label.image = None
        self.status_var.set("Изображения очищены")

if __name__ == "__main__":
    root = tkd
```
